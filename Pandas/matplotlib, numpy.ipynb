## 한 화면에 여러 개 그래프
- figure() : figure 객체 생성
- add_subplot(위치 및 개수) : 생성된 figure 객체에 subplot 추가
    - add_subplot(2, 1, 1) : 2x1 형태의 subplot, 두 개의 subplot 중 첫 번째
- add_subplot() 함수 호출 시 AxesSubplot 객체 생성
- 생성된 subplot 객체를 그래프 drawing 함수와 연결

```python
import matplotlib.pyplot as plt
%matplotlib inline

x = [i for i in range(100)]
y = [i**2 for i in x]

fig = plt.figure() # instance a figure object            
axe_01 = fig.add_subplot(2,1,1) # 1st subplot in 2x1
axe_02 = fig.add_subplot(2,1,2) # 2nd subplot in 2x2

axe_01.plot(x, y)               # plot a line graph
axe_02.bar(x, y)                # plot a bar graph
plt.show()
```

```python
fig = plt.figure() # instance a figure object            
axe_01 = fig.add_subplot(1,2,1) # 1st subplot in 2x1
axe_02 = fig.add_subplot(1,2,2) # 2nd subplot in 2x2

axe_01.plot(x, y)               # plot a line graph
axe_02.bar(x, y)                # plot a bar graph
plt.show()
```

## 그래프 라벨 및 범례 표시
- title(문자열) : 그래프 제목
- xlabel(문자열) : x축 라벨
- ylabel(문자열) : y축 라벨
- xlim((시작값, 끝값)) : x축 값 범위
- ylim((시작값, 끝값)) : y축 값 범위

```python
x = np.linspace(-5, 5, 1_000) # ndarray
y = np.power(x, 2)            # ndarray

plt.plot(x, y)   # plot a line with x, y
plt.title("2d graph")
plt.xlabel("x")  # x axe label
plt.ylabel('y')  # y axe label
plt.xlim((-5,5)) # range tuple
plt.ylim((0,25)) # range tuple
plt.grid()       # show a grid 
plt.show()       # show a whole graph
```

## histogram
- numpy.random.randn() : 임의의 표준 정규 분포 데이터 생성
- matplotlib의 hist() 함수 : 히스토그램 그리기
    - bin : 나누는 구간
    - cumulative : 누적 옵션
- grid(True) : 그리드 배경
- savefig("파일명") : 파일로 저장

```python
data = np.random.randn(1_000)

fig    = plt.figure()           # instance a figure object            
axe_01 = fig.add_subplot(121) # 1st subplot in 2x1
axe_02 = fig.add_subplot(122) # 2nd subplot in 2x2

axe_01.set_title("1st histogram")
axe_01.set_xlabel("x value")
axe_01.set_ylabel("y value")
axe_01.grid(True)
axe_01.hist(data, bins=30, density=True, color='b')

axe_02.set_title("1st histogram")
axe_02.set_xlabel("x value")
axe_02.set_ylabel("y value")
axe_02.grid(True)
axe_02.hist(data, bins=30, density=True, color='g', cumulative=True)

plt.subplots_adjust(wspace=.4)
plt.show()
```

## Horizontal bar

## 3D 그래프
- mpl_toolkits.mplot3d 모듈 추가
- scatter() : 점 그래프 그리기

```python
from mpl_toolkits.mplot3d import Axes3D

def randrange(n, min, max):
    return (max-min)*np.random.rand(n)+min

n = 100
fig = plt.figure(figsize=(12,6))
axe = fig.add_subplot(111, projection='3d')

for c, m, zl, zh in [('r', 'o', -50, -25), ('b', '^', -30, -5)]:
    xs = randrange(n, 23, 32)
    ys = randrange(n, 0, 100)
    zs = randrange(n, zl, zh)
    axe.scatter(xs, ys, zs, c=c, marker=m)

axe.set_xlabel('X Label')
axe.set_ylabel('Y Label')
axe.set_zlabel('Z Label')

plt.show()
```

# Activations function for Nueral Networks
- step function
- sigmoid function
- tanh(hperbolic tangent) function
- ReLU(Rectified Linear Unit or Ramp) function
- Leaky ReLU(Rectified Linear Unit or Ramp) function
- identity function
- softmax function
- logit(logistic probit) function

```python
# step function 
# perceptron 에서 사용하기 시작해서 유명해진 활성화 함수의 원조
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

def plot_function(x, y, fname, height = 3):  # common plot tool for several activation functions
    plt.figure(figsize=(12.5, height))
    plt.plot(x,y)
    plt.xlabel("x axis")
    plt.ylabel("y axis")
    plt.title(fname)
    plt.grid()
    plt.show()
    
def step_function(x):
    return np.where(x <= 0, 0, 1) # x = 0 if x <= 0 else 1

x = np.linspace(-5, 5)   # default num = 50
y = step_function(x)  

plot_function(x, y, "step function")
```

```python
# sigmoid function 
# 미분이 가능하기에 활성화 함수로 많이 선호

def sigmoid_function(x):
    return 1/(1+np.exp(-x)) # 미분하면 (1-y)y

x = np.linspace(-5, 5)    # default num = 50
y = sigmoid_function(x) 

plot_function(x, y, "sigmoid function")
```

```python
# tanh(hperbolic tangent) function 
# 시그모이드 함수(y는 0에서 1사이)와 비슷하지만 0을 중심으로 대칭(-1에서 1사이)

def tanh_function(x):
    return np.tanh(x)

x = np.linspace(-5, 5) # default num = 50
y = tanh_function(x)   

plot_function(x, y, "tanh function")
```

```python
# ReLU(Rectified Linear Unit or Ramp) function 
# 간단하면서도 층의 수가 많아져도 안정적으로 학습 가능하여 인기
# 출력층 이외의 함수에 활성화 함수로 많이 선호

def ReLU_function(x):
    return np.where(x <= 0, 0, x) # x = 0 if x <= 0 else x

x = np.linspace(-5, 5) 
y = ReLU_function(x)   

plot_function(x, y, "ReLU function")
```

```python
# Leaky ReLU(Rectified Linear Unit or Ramp) function 
# dying ReLU 현상(출력값이 0이 되면 더이상 학습이 진행되지 않는 뉴런 다수 발생) 방지

def Leaky_ReLU_function(x):
    return np.where(x <= 0, 0.01*x, x) # x = 0.01*x if x <= 0 else x

x = np.linspace(-5, 5)      
y = Leaky_ReLU_function(x)  

plot_function(x, y, "Leaky ReLU function", height=6)
```

```python
# identity function 
# 항등 함수는 y = x 형태로 입력 값이 곧 출력값인 형태로 
# 출력층의 활성화 함수로 소프트맥스와 함께 자주 사용
# 항등 함수는 출력 범위의 제한이 없고 연속적이기 때문에 
# 연속적인 수치를 예측하는 회귀(regression) 문제 다룰 때 적합

def identity_function(x):
    return x

x = np.linspace(-5, 5) # numpy.ndarray, (50,) --> row vector or (1,50) matrix
y = identity_function(x)  # numpy.ndarray, (50,) --> column vector or (50,1) matrix

plot_function(x, y, "identity function")
```

```python
# softmax function
# 소프트맥스 함수는 분류문제를 다룰 때 출력층의 활성화 함수로 적합하여 많이 선호
# 0 < y < 1  : 소프트맥스 함수를 통과한 값은 0에서 1 사이란 뜻
# 이 값은 각 뉴런이 특정한 범주로 분류될 확률을 나타낼 때 사용

def softmax_function(x):
    return np.exp(x) / np.sum(np.exp(x))

x = np.linspace(-5, 5, 100) 
y = softmax_function(x)  

plot_function(x, y, "softmax function", height= 3)
```

```python
# logit(logistic probit) function
# logit과 sigmoid는 서로 역함수 관계이고
# 2개 클래스를 대상으로 정의하던 logit을 K개의 클래스를 대상으로 일반화하면 
# softmax함수가 유도(derived)됩니다.
# softmax함수에서 K를 2로 놓으면 sigmoid함수로 환원이 되고,
# 반대로 sigmoid함수를 K개의 클래스를 대상으로 일반화하면 softmax함수가 유도됩니다
# 
def logit_function(x):
    return np.log(x/(1-x))

x = np.linspace(-5, 5, 1000) 
y = logit_function(x)  

plot_function(x, y, "logit function")
```
