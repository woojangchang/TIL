# 이미지 반전 처리와 회전 처리
- 이미지 수가 많을 수록 학습률이 올라가므로 필요함

## 좌우 반전과 상하 반전
- `cv2.flip()` 사용
    - 0 : 상하 반전
    - 1 : 좌우 반전

```python
import matplotlib.pyplot as plt
import cv2

img = cv2.imread('../datasets/test.jpg')

plt.figure(figsize=(10, 6))

plt.subplot(1, 3, 1)
plt.axis('off')
plt.title('original image')
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))

# 이미지 좌우 반전
plt.subplot(1, 3, 2)
plt.axis('off')
plt.title('left and right reversed image')
img2 = cv2.flip(img, 1)
plt.imshow(cv2.cvtColor(img2, cv2.COLOR_BGR2RGB))

# 이미지 상하 반전
plt.subplot(1, 3, 3)
plt.axis('off')
plt.title('upside down image')
img3 = cv2.flip(img, 0)
plt.imshow(cv2.cvtColor(img3, cv2.COLOR_BGR2RGB))
plt.show()
```

## 이미지 회전
- `matrix = cv2.getRotationMatrix2D(중심점, 회전 각도, 확대 비율)` : 회전 변환 행렬 생성
- `cv2.warpAffine(img, matrix, size)` : 회전 변환 행렬에 따라 이미지를 회전

```python
import matplotlib.pyplot as plt
import cv2

img = cv2.imread('../datasets/test.jpg')

# 이미지 크기 추출
h, w, colors = img.shape
size = (w, h)

# 이미지 중심 위치 추출
center = (w // 2, h // 2)

# 회전 변환 행렬 만들기
angle = 45
scale = 1.0

matrix = cv2.getRotationMatrix2D(center, angle, scale)

# 변환하기
img2 = cv2.warpAffine(img, matrix, size)

plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))

plt.subplot(1, 2, 2)
plt.imshow(cv2.cvtColor(img2, cv2.COLOR_BGR2RGB))
plt.show()
```
