## 이미지 불러오기

```python
import urllib.request as req
url = 'http://uta.pw/shodou/img/28/214.png'
req.urlretrieve(url, '../datasets/opencv_test.png')

img = cv2.imread('../datasets/opencv_test.png')
print(img)
```

### 이미지 인라인 출력
- OpenCV : BGR
- matplotlib : RGB
- 그러므로 `cv2.cvtColor(img, cv2.COLOR_BGR2RGB)`로 색공간을 변환

```python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline

img = cv2.imread('../datasets/test.jpg')
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
# 색공간 변환을 하지 않을 경우
plt.imshow(img)
plt.show()
```

```python
plt.axis('off')
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.show()
```

### 이미지 저장
- BMP, PPM, PBM, PGM, JPEG, PNG, TIFF, WebP 등 대부분 이미지 형식 지원

```python
import cv2

img = cv2.imread('../datasets/test.jpg')

cv2.imwrite('../datasets/out.png', img)
```

### 이미지 크기 변경
- `cv2.resize(img, (width, height))` 사용

```python
im2 = cv2.resize(img, (600, 300))

plt.imshow(cv2.cvtColor(im2, cv2.COLOR_BGR2RGB))
plt.show()
```

### 이미지 자르기
- `배열[y1:y2, x1:x2]`

```python
im2 = img[150:450, 150:450]

im2 = cv2.resize(im2, (400, 400))
plt.imshow(cv2.cvtColor(im2, cv2.COLOR_BGR2RGB))
plt.show()
```

## 헷갈리지 말아야 할 것
- 슬라이싱은 `img[y1:y2, x1:x2]`로 numpy와 같이 행-열(y-x) 순서이지만
- 리사이징은 `(width, height)`로 열-행(x-y) 순서
