# Cross Entropy Error
- 분류 평가 지표 중 하나  
- 이진 분류  
$CE_i = -y_i \log(\hat{y_i}) - (1-y_i)\log{(1-\hat{y_i})}$  
$y_i \in \{0, 1\} \,, \hat y_i=\text{sigmoid}(w_ix+b_i) \in [0, 1]$
  - $y_i=0 ; CE=-\log (1-\hat y_i)$
  - $y_i=1 ; CE=-\log (\hat y_i)$
  - $y_i=0, \hat y_i=0$ 또는 $y_i=1, \hat y_i=1$이면 $CE=0$
  - $y_i=0, \hat y_i=1$ 또는 $y_i=1, \hat y_i=0$이면 $CE=\infty$  
  
- 다중 분류  
$CE_i = -y_i \log(\hat{y_i})$

## Cross Entropy

```python
import numpy as np

def CE(y, y_hat):
  if y == 0:
    return -np.log(1-y_hat)
  elif y == 1:
    return -np.log(y_hat)
  else:
    return -y * np.log(y_hat) - (1-y) * np.log(1-y_hat)
```

```python
CE(1, 1)
```

```python
CE(0, 0)
```

```python
CE(1, 0.0001)
```

```python
CE(0, 0.9999)
```

## Information Theory
- Information Gain (정보 이득량)
  - 자주 발생하지 않는 사건은 자주 발생하는 사건보다 전달하는 정보량이 많음
  - $I(x) = -\log(P(x))$
  - $P(x)$ : 사건 x가 발생할 확률

```python
A = 0.9
B = 0.5
C = 0.1

print(f'{-np.log(A):.3f}, {-np.log(B):.3f}, {-np.log(C):.3f}')
```

## Entropy
- 불확실성의 정도
  - $E(-\log(P(x)))$
- 확률변수의 평균 정보량(기댓값)
  - $-\sum (p(x) \log(p(x)))$
- 불확실성(Entropy)이 낮으면 분류 정확도가 높아짐

```python
def entropy(*iters):
  e = 0
  for p in iters:
    e -= p * np.log(p)
  return e
```

```python
# 승률이 비슷한 두 팀의 엔트로피
P1 = 0.5
P2 = 0.5

entropy(P1, P2)
```

```python
# 승률 차이가 큰 두 팀의 엔트로피
P1 = 0.999
P2 = 0.001

entropy(P1, P2)
```

```python
import matplotlib.pyplot as plt

xs = np.linspace(0.001, 0.999, 999)
ys = entropy(xs, 1-xs)
plt.plot(xs, ys)
plt.title('Entropy')
plt.show()
```
