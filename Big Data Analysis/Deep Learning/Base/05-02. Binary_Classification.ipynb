# Binary Classification
- default YES : 연체, NO : 상환

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
import pandas as pd

DF = pd.read_csv('https://raw.githubusercontent.com/rusita-ai/pyData/master/Default.csv')

DF.info()
```

## EDA

### 빈도 분석

```python
DF.default.value_counts()
```

### 분포 시각화

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(9, 6))
plt.boxplot([DF[DF.default == 'No'].balance,
            DF[DF.default == 'Yes'].balance],
            labels=['No', 'Yes'])
plt.show()
```

## Data Preprocessing

### Standardization

```python
X = DF[['balance']]
y = DF['default']
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_Scaled = scaler.fit_transform(X)

X_Scaled[:5]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_Scaled, y, test_size=0.3, random_state=2045)
```

## Modeling

```python
from sklearn.linear_model import LogisticRegression

Model_lr = LogisticRegression()
Model_lr.fit(X_train, y_train)
```

```python
y_hat = Model_lr.predict(X_test)
y_hat
```

## Validation

### Accuracy

```python
Model_lr.score(X_train, y_train)
```

```python
Model_lr.score(X_test, y_test)
```

### Confusion Matrix

- No(상환) 기준

```python
from sklearn.metrics import confusion_matrix

confusion_matrix(y_test, y_hat)
```

- Yes(연체) 기준

```python
confusion_matrix(y_test, y_hat, labels=['Yes', 'No'])
```

### Accuarcy, Precision, Recall - No

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score

print(accuracy_score(y_test, y_hat))
print(precision_score(y_test, y_hat, pos_label='No'))
print(recall_score(y_test, y_hat, pos_label='No'))
```

### Accuarcy, Precision, Recall - Yes

```python
print(accuracy_score(y_test, y_hat))
print(precision_score(y_test, y_hat, pos_label='Yes'))
print(recall_score(y_test, y_hat, pos_label='Yes'))
```

### F1 Score

```python
from sklearn.metrics import f1_score

print(f1_score(y_test, y_hat, pos_label='No'))
print(f1_score(y_test, y_hat, pos_label='Yes'))
```

### Classification Report

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, y_hat,
                            target_names=['No', 'Yes'],
                            digits=5))
```
