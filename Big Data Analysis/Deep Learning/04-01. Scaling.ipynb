```python
import warnings
warnings.filterwarnings('ignore')
```

```python
import seaborn as sns

DF = sns.load_dataset('mpg')
```

```python
DF.info()
```

```python
DF.head()
```

```python
X = DF[['weight']]
y = DF['mpg']
```

```python
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(9, 6))
sns.regplot(x=X, y=y)
plt.xlabel('weight_Without_Scaling')
plt.show()
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler1 = MinMaxScaler()
X_Norm = scaler1.fit_transform(X)
```

```python
fig = plt.figure(figsize=(9, 6))
sns.regplot(x=X_Norm, y=y)
plt.xlabel('weight_Witht_Normalization')
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler

scaler2 = StandardScaler()
X_Stan = scaler2.fit_transform(X)
```

```python
fig = plt.figure(figsize=(9, 6))
sns.regplot(x=X_Stan, y=y)
plt.xlabel('weight_Witht_Standardization')
plt.show()
```
