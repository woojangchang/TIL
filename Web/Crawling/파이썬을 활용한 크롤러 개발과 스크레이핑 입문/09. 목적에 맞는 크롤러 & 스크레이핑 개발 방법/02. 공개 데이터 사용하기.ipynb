````python
import urllib
import requests
from bs4 import BeautifulSoup
import json

# URL 관련 정보 선언
serviceKey = '서비스키'
end_point = 'http://apis.data.go.kr/B552584/ArpltnInforInqireSvc/getMsrstnAcctoRltmMesureDnsty?serviceKey=' + serviceKey
default_options = {
    'numOfRows':1,
    'pageNo':1,
    'dataTerm':'Daily',
    'ver':1.3
}
````

# xml

````python
def get_dust_info(region, returnType='xml'):
    """특정 지역의 미세먼지 수치 추출하기"""
    
    # URL 생성
    default_options['stationName'] = region
    default_options['returnType'] = returnType
    params = urllib.parse.urlencode(default_options)
    url = end_point + '&' + params
    
    # 요청
    r = requests.get(url)
    txt = r.text
    
    # 필요한 정보를 스크레이핑
    if returnType == 'xml':
        soup = BeautifulSoup(txt, 'html.parser')
        date = soup.select('item > datatime')[0].text
        pm_10 = soup.select('item > pm10value')[0].text
        pm_25 = soup.select('item > pm25value')[0].text
    else:
        soup = json.loads(txt)
        data = soup['response']['body']['items'][0]
        date = data['dataTime']
        pm_10 = data['pm10Value']
        pm_25 = data['pm25Value']
    
    return {
        "region":region,
        "datatime":date,
        "pm_10":pm_10,
        "pm_25":pm_25
    }
````

````python
if __name__ == '__main__':
    output = get_dust_info('관악구')
    print(json.dumps(output, ensure_ascii=False))
````
