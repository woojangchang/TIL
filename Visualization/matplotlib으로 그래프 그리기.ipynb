## 여러 그래프 그리기
- `plt.plot(x1, y1); plt.plot(x2, y2)`
- `plt.plot(x1, y1 [, fmt1], x2, y2 [,fmt2])`

```python
import numpy as np

x = np.arange(-4.5, 5, 0.5)
y1 = 2 * x ** 2
y2 = 5 * x + 30
y3 = 4 * x ** 2 + 10
```

```python
plt.plot(x, y1)
plt.plot(x, y2)
plt.plot(x, y3)
plt.show()
```

```python
plt.plot(x, y1, x, y2, x, y3)
plt.show()
```

```python
# 새로운 그래프 창을 생성해서 그래프 그리기
plt.plot(x, y1)

plt.figure() # 새로운 그래프 창 생성

plt.plot(x, y2)
plt.show()
```

```python
# 그래프 창의 번호를 명시적으로 지정하여 해당 창에 그래프 그리기
import numpy as np

# 데이터 생성
x = np.arange(-5, 5, 0.1)
y1 = x**2 - 2
y2 = 20 * np.cos(x) ** 2

plt.figure(1) # 1번 그래프 창 생성
plt.plot(x, y1) # 지정된 그래프 창에 그래프를 그림

plt.figure(2) # 2번 그래프 창 생성
plt.plot(x, y2)

plt.figure(1) # 이미 생성된 1번 그래프 창 지정
plt.plot(x, y2)

plt.figure(2) # 이미 생성된 2번 그래프 창 지정
plt.clf() # 2번 그래프 창에 그려진 모든 그래프를 지움
plt.plot(x, y1)

plt.show()
```

```python
import numpy as np

# 데이터 생성
x = np.arange(0, 10, 0.1)
y1 = 0.3 * (x-5) ** 2 + 1
y2 = -1.5 * x + 3
y3 = np.sin(x) ** 2
y4 = 10 * np.exp(-x) + 1

# 2 x 2 행렬로 이뤄진 하위 그래프에서 p에 따라 위치를 지정
plt.subplot(2, 2, 1)
plt.plot(x, y1)

plt.subplot(2, 2, 2)
plt.plot(x, y2)

plt.subplot(2, 2, 3)
plt.plot(x, y3)

plt.subplot(2, 2, 4)
plt.plot(x, y4)

plt.show()
```

## 출력 형식 지정
- `fmt = '[color][line_style][marker]'`

```python
# 데이터 생성
x = np.arange(0, 5, 1)
y1 = x
y2 = x + 1
y3 = x + 2
y4 = x + 3
```

```python
# 옵션 지정 없이 그래프 생성
plt.plot(x, y1, x, y2, x, y3, x, y4)
plt.show()
```

### 컬러 지정
- 컬러 지정 약어
    - b, g, r, c(청녹색), m(자홍색), y, k, w

```python
plt.plot(x, y1, 'm', x, y2, 'y', x, y3, 'k', x, y4, 'c')
plt.show()
```

### 선 스타일 지정
- 선 스타일 약어
    - -, --, :, -.

```python
plt.plot(x, y1, '-', x, y2, '--', x, y3, ':', x, y4, '-.')
plt.show()
```

### 마커 표시
- 마커 지정 약어
    - o, ^, v, <, >, s(사각형), p(오각형), h(육각형1), H(육각형2), \*, +, x, X(채워진 x), D(다이아몬드), d(얇은 다이아몬드)

```python
plt.plot(x, y1, 'H', x, y2, '^', x, y3, 's', x, y4, 'd')
plt.show()
```

### 혼합 지정

```python
plt.plot(x, y1, '>--r', x, y2, 'h-g', x, y3, 'D:b', x, y4, '-.Xc')
plt.show()
```

## 산점도
- `plt.scatter(x, y, [,s=size_n, c=colors, marker='marker_string', alpha=alpha_f])`
    - s : 마커의 크기
    - c : 색
    - marker : 모양
    - alpha : 투명도(0 완전 투명, 1 완전 불투명)

```python
import matplotlib.pyplot as plt

height = [165, 177, 160, 180, 185, 155, 172]
weight = [62, 67, 55, 74, 90, 43, 64]

plt.scatter(height, weight)
plt.xlabel('Height(cm)')
plt.ylabel('Weight(kg)')
plt.title('Height & Weight')
plt.grid(True)
```

```python
# 우리나라 주요 도시의 인구 밀도 시각화 예
import numpy as np

city = ['서울', '인천', '대전', '대구', '울산', '부산', '광주']

lat = [37.56, 37.45, 36.35, 35.87, 35.53, 35.18, 35.16] # 위도
lon = [126.97, 126.70, 127.38, 128.60, 129.31, 129.07, 126.85] # 경도

# 인구 밀도(명/km^2) : 2020년 통계청 자료
pop_den = [15865, 2770, 2780, 2738, 1073, 4342, 2969]

size = np.array(pop_den) * 0.2 # 마커 크기 지정
colors = ['r', 'g', 'b', 'c', 'm', 'k', 'y'] # 마커 색 지정

plt.scatter(lon, lat, s=size, c=colors, alpha=0.5)
plt.xlabel('경도(longitude)')
plt.ylabel('위도(latitude)')
plt.title('지역별 인구 밀도(2020)')

# 위도 경도에 맞게 도시 이름 출력
for x, y, name in zip(lon, lat, city):
    plt.text(x-0.08, y-0.04, name)
    
plt.show()
```

# 막대 그래프
- `plt.bar(x, height [,width=width_f, color=colors, tick_label=tick_labels, align='center'(기본) 혹은 'edge', label=labels])`

```python
# 데이터 생성
member_IDs = ['m_01', 'm_02', 'm_03', 'm_04']
before_ex = [27, 35, 40, 33]
after_ex = [30, 38, 42, 37]
```

```python
import matplotlib.pyplot as plt
import numpy as np

n_data = len(member_IDs)
index = np.arange(n_data)
plt.bar(index, before_ex) # x=index, height=before_ex
plt.show()
```

```python
# x축 라벨 지정
plt.bar(index, before_ex, tick_label=member_IDs)
plt.show()
```

```python
# 그래프 색 변경
colors = ['r', 'g', 'b', 'm']
plt.bar(index, before_ex, color=colors, tick_label=member_IDs)
plt.show()
```

```python
# 폭 변경
plt.bar(index, before_ex, tick_label=member_IDs, width=0.6)
plt.show()
```

```python
# 가로 막대 그래프
colors = ['r', 'g', 'b', 'm']
plt.barh(index, before_ex, color=colors, tick_label=member_IDs)
plt.show()
```

- `bar()`의 `align` 옵션은 `edge`로 지정해 막대 그래프를 한쪽으로 치우치게 함
- `width` 옵션은 두 개의 막대 그래프가 들어갈 수 있도록 여유 있게 0.4로 지정
- `after_ex` 데이터를 이용해 막대 그래프를 그릴 때 `before_ex` 데이터를 이용해 그린 막대 그래프와 겹치지 않게 x좌표에서 막대 그래프의 시작 위치를 막대 그래프의 두께(0.4)만큼 오른쪽으로 이동
- 범례로 두 데이터를 구분하기 위해 `label` 옵션에 각각 문자열을 지정, 이후에 `legend()` 실행
- 두 개의 데이터를 그린 경우에는 `tick_label` 옵션을 이요해 tick 라벨을 변경할 수 없으므로 `xticks(tick_위치, tick_label)`를 이용해 x축의 tick 라벨을 붙임

```python
barWidth = 0.4
plt.bar(index, before_ex, color='c', align='edge', width=barWidth, label='before')
plt.bar(index+barWidth, after_ex, color='m', align='edge', width=barWidth, label='after')

plt.xticks(index+barWidth, member_IDs)
plt.legend()
plt.xlabel('회원 ID')
plt.ylabel('윗몸일으키기 횟수')
plt.title('운동 시작 전과 후의 근지구력(복근) 변화 비교')
plt.show()
```

# 파이 그래프
- `plt.pie(x [,labels=label_seq, autopct='비율 표시 형식(ex: %0.1f)', shadow=False(기본값) 혹은 True, explode=explode_seq, counterclock=True(기본) 혹은 False, startangle=각도(기본값 0)])`

```python
fruit = ['사과', '바나나', '딸기', '오렌지', '포도']
result = [7, 6, 3, 2, 2]
```

```python
plt.pie(result)
plt.show()
```

## 라벨, 비율 추가

```python
plt.pie(result, labels=fruit, autopct='%.1f%%')
plt.show()
```

```python
# x축 기준 각도 90도에서 시작해서 시계방향으로 설정
plt.pie(result, labels=fruit, autopct='%.1f%%', startangle=90, counterclock=False)
plt.show()
```

```python
# 특정 요소 강조
explode_value = (0.1, 0, 0, 0, 0)

plt.pie(result, labels=fruit, autopct='%.1f%%', startangle=90, counterclock=False,
        explode=explode_value, shadow=True)
plt.show()
```
